```python
from astropy.table import Table
table = Table.read("global_CCl4_MM.dat", format = "ascii")
```

```python
df3 = table.to_pandas()
df3.columns
```

```python
df3 = df3[["CCl4ottoyr", "CCl4ottoGLm", "CCl4ottoGLsd"]]
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.errorbar(x=df3["CCl4ottoyr"], y=df3["CCl4ottoGLm"], yerr=df3["CCl4ottoGLsd"], capsize=5)
plt.show()
```

```python
from scipy.optimize import curve_fit
def linear_fit(x, m, b):
    return m*x + b
cols = ["CCl4ottoyr", "CCl4ottoGLm", "CCl4ottoGLsd"]
mask = np.isfinite(df3[cols]).all(axis=1)
x = df3.loc[mask, "CCl4ottoyr"].values
y = df3.loc[mask, "CCl4ottoGLm"].values
yerr = df3.loc[mask, "CCl4ottoGLsd"].values
popt, pcov = curve_fit(linear_fit, xdata=x, ydata=y)
m_fit, b_fit = popt
plt.errorbar(x=x, y=y, yerr=yerr, capsize=5)
plt.plot(x, linear_fit(x, m_fit, b_fit), color="gold")
plt.xlabel("CCl4ottoyr")
plt.ylabel("CCl4ottoGLm")
plt.show()
residuals = y - linear_fit(x, m_fit, b_fit)
chi2 = np.sum(residuals**2)
dof = len(x) - len(popt)
chi2_squared = chi2 / dof
print(f"Reduced chi-squared: {chi2_squared:.2f}")
```

```python
# 3.3.6) The model fits well, it has a reduced chi-squared of only 0.18
```
