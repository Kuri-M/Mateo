```python
#1a) Transit, RV, Microlensing, Imaging, Transit Timing Variations, Eclipse Timing Variations, Disk Kinematics
#1b) Transit has most with 388, Disk Kinematics has least with 1
#1c) Transit: Planet passes in front of star, Radial Velocity: Doppler shift of star shows planet, Microlensing: Gravitational Magnified light has a dip in it from a planet in the system, Imaging: Direct photo, TTV: A transit dip is irregular and can only be explained by another exoplanet, ETV: Eclipses of stars have irregularities which are due to an exoplanet, Disk Kinematics: Protoplanetry disk disturbances
```

```python
from matplotlib.lines import Line2D 
colors = []
method_to_color = {}
for thing in df2["discoverymethod"]:
    if thing == "Transit":
        c = "darkslategrey"
    elif thing == "Radial Velocity":
        c = "indianred"
    elif thing == "Microlensing":
        c = "deepskyblue"
    elif thing == "Imaging":
        c = "limegreen"
    elif thing == "Transit Timing Variations":
        c = "salmon"
    elif thing == "Eclipse Timing Variations":
        c = "chartreuse"
    else:
        c = "palegreen"
    colors.append(c)
    method_to_color.setdefault(thing, c)
handles = [
    Line2D([0], [0], marker='o', linestyle='', color=col, markersize=8)
    for col in method_to_color.values()
]
plt.figure(figsize=(8,6))
plt.scatter(x="pl_bmassj", y="pl_eqt", c=colors, data=df2)
plt.title("Mass vs. Temp in Exoplanets by Discovery Method")
plt.xlabel("Planet Mass (In Jupiter Masses)")
plt.ylabel("Equilibrium Temperature in Kelvin")
labels = list(method_to_color.keys())
plt.legend(handles, labels, title="Discovery Method")
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x="pl_bmassj", y="pl_eqt", c=colors, data=df2)
plt.title("Mass vs. Temp in Exoplanets by Discovery Method")
plt.xlabel("Planet Mass (In Jupiter Masses)")
plt.ylabel("Equilibrium Temperature in Kelvin")
plt.xscale("log")
labels = list(method_to_color.keys())
plt.legend(handles, labels, title="Discovery Method")
plt.show()
```

```python
#4) Exoplanets found by the imaging method tend to be much larger
i = 2
for index, row in df2.iterrows():
    if row["pl_bmassj"] > 5 and row["pl_eqt"] < df2["pl_eqt"][i]:
        i = index
print(df2["pl_name"][i])
print(df2["pl_eqt"][i])
print(df2["pl_bmassj"][i])
print(df2["hostname"][i])
print(df2["st_spectype"][i])
print(i)
```

```python
#5) The planet with the lowest equilibrium temp and a mass greater than x5 jupiters is eps Ind A b.
#6) Name: eps Ind A b, index: 566, equilibrium temperature: 275, mass in Jupiter Masses: 6.31, The star is call eps Ind A and is of type K5 V.
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x="pl_bmassj", y="pl_eqt", c=colors, data=df2)
plt.scatter(x=6.31, y=275.0, c="aqua", s=200, label="eps Ind A b")
coldest_handle = plt.scatter(x=6.31, y=275.0, c="aqua", s=200, label="eps Ind A b")
plt.title("Mass vs. Temp in Exoplanets by Discovery Method")
plt.xlabel("Planet Mass (In Jupiter Masses)")
plt.ylabel("Equilibrium Temperature in Kelvin")
plt.xscale("log")
labels = list(method_to_color.keys())
plt.legend(handles + [coldest_handle], labels + ["eps Ind A b"], title="Discovery Method")
plt.show()
```
