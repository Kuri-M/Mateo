```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Creating our Sin wave and Subplots:
def make_sin_wave(x, A, w):
    return A*np.sin(w*x)
x = np.linspace(0, 2*np.pi, 1000)
fig, axis = plt.subplots(nrows=1, ncols=3, figsize=(25, 5))
amplitude = np.array([0.5, 1, 1.5, 2,2.5])
frequency = np.array([1, 2, 3, 4, 5])
for A_val, w_val in zip(amplitude, frequency):
    y1 = make_sin_wave(x, A_val, 1)
    axis[0].plot(x,y1, label = "A"),
    y2 = make_sin_wave(x, 1, w_val)
    axis[1].plot(x,y2, label = "w")
    y3 = make_sin_wave(x, A_val, w_val)
    axis[2].plot(x,y3, label = "A, w")
    
# Labeling for Axis 1,2,3:

axis[0].set_title("Sin Wave Changing in A"), axis[0].set_ylabel("y1"), axis[0].set_xlabel("0-2pi"), axis[0].legend()
axis[1].set_title("Sin Wave Changing in w"), axis[1].set_ylabel("y2"), axis[1].set_xlabel("0-2pi"), axis[1].legend()
axis[2].set_title("Sin Wave Changing in A and w"), axis[2].set_ylabel("y3"), axis[2].set_xlabel("0-2pi"), axis[2].legend()

plt.show()
```
