```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = np.loadtxt("star_data.txt",skiprows=1)

print(data.shape)

# Theres 2 columns and 2642 rows 

mag = magnitude = data[:,0]
temp = temperature = data[:,1]

print(f"Max_Mag: {magnitude.max()}")
print(f"Min_Mag: {magnitude.min()}")

print(f"Max_Temp: {temperature.max()}")
print(f"Min_Temp: {temperature.min()}")

# The smallest magnitude is the brightest numerically, while the largest magnitude is the most dim numerically 
```

```python
log_magnitude = np.log(mag)
log_temperature = np.log(temp)

fig = plt.figure(figsize=(20,8))
y = log_magnitude
x = log_temperature

plt.scatter(x,y, label = "Stars")
plt.gca().invert_xaxis() 
plt.gca().invert_yaxis()

# Max Mag Point - Dimmist Star
max_mag = np.argmax(mag)
max_mag_val = mag[max_mag]
temp_at_maxmag = temp[max_mag]
plt.scatter(
    np.log(temp_at_maxmag), 
    np.log(max_mag_val), 
    color = "purple", marker = "*", s=200, label = "Dimmest")


# Min Mag Point - Brightest Star
min_mag = np.argmin(mag)
min_mag_val = mag[min_mag]
temp_at_minmag = temp[min_mag]
plt.scatter(
    np.log(temp_at_minmag),
    np.log(min_mag_val),
    color = "black", marker = "*", s=200, label = "Brightest")


# Max temp Point - Hottest Star
max_temp = np.argmax(temp)
max_temp_val = temp[max_temp]
mag_at_maxtemp = mag[max_temp]
plt.scatter(np.log(max_temp_val),
            np.log(mag_at_maxtemp),
            color = "green", marker = "*", s=200, label = "Hottest")


# Min temp Point - Coolest Star
min_temp = np.argmin(temp)
min_temp_val = temp[min_temp]
mag_at_mintemp = mag[min_temp]
plt.scatter(np.log(min_temp_val),
            np.log(mag_at_mintemp),
            color = "orange", marker = "*", s=200, label = "Coolest")


# Labeling 
plt.xlabel("Temperature (logK)")
plt.ylabel("Magnitude")
plt.title("Hertzsprung-Russel Diagram")
plt.legend()
```
